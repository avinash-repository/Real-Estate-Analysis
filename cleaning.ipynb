```python
import pandas as pd
import random

# Load your CSV file
file_path = 'E:\\Data Analyst\\Project\\Scraped Dataset\\Cleaned_Real_estate.csv'
df = pd.read_csv(file_path)

# Filter for rows with Type 'Commercial' and Address containing 'New York' (ignoring case)
commercial_ny = df[(df['Type'] == 'Commercial') & (df['Address'].str.contains('New York', case=False, na=False))]

# Keep the first 30 New York addresses unchanged
ny_keep = commercial_ny.head(30)

# Remaining rows after the first 30 that need to be updated
remaining_ny = commercial_ny.iloc[30:]

# Define a list of different cities/states to distribute the remaining addresses
locations = [
    ('California', 'Los Angeles', 90001),
    ('California', 'San Francisco', 94101),
    ('Texas', 'Austin', 73301),
    ('Florida', 'Miami', 33101),
    ('Utah', 'Salt Lake City', 84101),
    ('Illinois', 'Chicago', 60601),
    ('Ohio', 'Columbus', 43201),
    ('Nevada', 'Las Vegas', 89101),
    ('Washington', 'Seattle', 98101),
    ('Arizona', 'Phoenix', 85001),
    ('Georgia', 'Atlanta', 30301),
    ('North Carolina', 'Charlotte', 28202)
]

# Function to generate random street addresses
def random_street():
    street_number = random.randint(100, 999)
    street_name = random.choice(['Main St', 'Maple Ave', 'Oak St', 'Pine St', 'Elm St', 'Cedar Blvd', 'Sunset Blvd'])
    return f'{street_number} {street_name}'

# Loop through the remaining New York commercial addresses
for i, row in remaining_ny.iterrows():
    state, city, pincode = random.choice(locations)
    random_address = random_street()  # Generate a random street address
    # Update the Address and Pincode directly in the main dataframe (without dropping rows)
    df.loc[row.name, 'Address'] = f'{random_address}, {city}, {state}'
    df.loc[row.name, 'Pincode'] = pincode
    # Leave the Type unchanged (still "Commercial")

# Combine the unchanged rows with the modified dataframe (without dropping any rows)
# Ensure no rows from the original dataframe are dropped.
df_updated = df.copy()

# Save the updated dataframe to a new CSV file
output_file_path = 'E:\\Data Analyst\\Project\\Scraped Dataset\\Updated_Real_estate.csv'
df_updated.to_csv(output_file_path, index=False)

print(f'Updated file saved at {output_file_path}')
```
